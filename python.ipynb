```python
def add(a,b):
    x=a+b
    return x
def sub(a,b):
    x=a-b
    return x
def mul(a,b):
    x=a*b
    return x
def div(a,b):
    x=a/b
    return x
def modulus(a,b):
    x=(a%b)
    return x
def exponent(a,b):
    x=a**b
    return x
print(add(34,67))
print(sub(34,4))
print(mul(4,9))
print(div(40,2))
print(modulus(34,34))
print(exponent(2,2))
```

```python
# without function for banking sector
loan_amount=500000
intrest=loan_amount*0.1
print(intrest)

loan_amount=700000
intrest=loan_amount*0.1
print(intrest)
```

```python
# with function
def calculate_intrest(loan_amount):
    intrest=loan_amount*0.1
    return intrest

print(calculate_intrest(500000))
print(calculate_intrest(700000))
print(calculate_intrest(100000))
```

```python
# Example for manufacturing/

def machine_status(temp):
    if temp>100:
       return "shutdown"
    elif temp>80:
       return "Warning"
    else:
       return "Normal"
    
print(machine_status(75))
print(machine_status(110))
```
