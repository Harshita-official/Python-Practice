```python
class Student:
    def __init__(self,name,grade):
        self.name= name #Attribute
        self.grade= grade #Attribute
    def get_grade(self): #Method
        return f"{self.name} is in grade {self.grade}"    
```

```python
# Object creation
Harshita=Student("Harshita",12)
print(Harshita.get_grade())
```

```python
Sayli=Student("Sayli",12)
print(Sayli.get_grade())
```

```python
Sumit=Student("Sumit",12)
print(Sumit.get_grade())
```

```python
# with if else
class St:
    def __init__(self,name,course,number,attendance):
        self.name=name
        self.course=course
        self.number=number
        self.attendance=attendance
    def topics(self):
        if self.attendance.lower()=="present":
            return f" If  {self.name} is present today for the {self.course} class so she can be conntacted at {self.number}"
        else:
            return f"{self.name} is Absent today for the {self.course} class so she can be conntacted."

Harshita=St("Harshita","Data Analytics","8799820207","Absent")
print(Harshita.topics())            
```

```python
# Example for banking indusrty
class BankAccount:
    def __init__(self,customer_name,balance):
       self.customer_name=customer_name
       self.balance=-balance

    def deposit(self,amount):
        self.balance += amount
        print(f"₹{amount}deposited.")

    def withdraw(self,amount):
        if amount<=self.balance:
            self.balance -= amount
            print(f"₹{amount}withdrawn.")
        else:
           print("Insufficiant Balance")

    def check_balance(self):
        #  return self.balance
        print(f"Currente Balance:₹{self.balance}")

account=BankAccount("Harshita",10)

account.deposit(5)
account.withdraw(4)
account.check_balance()
```

```python
class Machine:

    def __init__(self,machine_id):
        self.machine_id=machine_id
        self.production_count = 0
    
    def produce(self,quantity):
        self.production_count += quantity
        print(f"Produced {quantity} units")
    
    def machine_status(self):
        print(f"Machine{self.machine_id} is Running")

    def total_production(self):
        print(f"Total Production: {self.production_count}")

machine = Machine("M101")

machine.produce(100)
machine.produce(130)
machine.total_production()
```
